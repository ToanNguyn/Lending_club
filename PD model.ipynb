Independent variables that significantly predict borrower default are identified by checking the statistical significance of their coefficients, commonly using p-values, and only significant predictors are kept in the final model.

Variable removal rules:

- Keep all dummies if all are significant; remove all if none are significant.

- Each original variable corresponds to multiple dummies.

- If any dummy is significant, keep all dummies for that original variable.

- A p-value < 0.05 indicates statistical significance.
